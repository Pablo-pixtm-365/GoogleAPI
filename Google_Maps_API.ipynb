A. Use the [*Nearby Seach*](https://developers.google.com/places/web-service/search) Method to find all the restaurant near to the UPY (20.988459, -89.736768) around 2km. In order to fulfill this item you need to:
* Create a *search_nearby_places* function with the appropiate parameters. It must return a JSON with the results. Places must be ranked by distance.
* Create a table with the following columns: Name, Place Id, Rating, Place Types, Total of User Ratings.

```python
"""import the libraries to use"""
import pandas as pd 
import requests
import json
```

```python
def search_nearby_places(key, location, radius, rankby=None):
    """In this function return a json with the restaurant that are to 2 km around of UPY 
    the arguments in this function is the key-> our API key, location->lat and lon of the UPY and rankby->None"""
    
    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?types=restaurant'
    
    parameters = {'key':key, 'location':location, 'radius':radius, 'rankby':rankby}
 
    
    res = requests.get(url, params=parameters)
    results = json.loads(res.content)
    return results
```

```python
places_json = search_nearby_places(key, location, radius, rankby)
places_json
```

```python
df = pd.read_json(json.dumps(places_json['results']), orient = 'records')
df.head()
```

B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
def find_place(key, input_1, inputtype, locationbias=None):
    """This function return the ID of a place, if you want you can introduce lat and lon 
    the arguments: key->our API key, input_1->name our place, intputtype-> the type of input
    that we want in this case text, locationbias-> lat and long"""
    
    url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json'
    if(locationbias!=None):
        parameters = {
            'key':key,
            'input':input_1,
            'inputtype':inputtype,
            'locationbias':str(locationbias)
        }
    else:
        parameters = {
            'key':key,
            'input':input_1,
            'inputtype':inputtype,
            'locationbias':None
        }
    res = requests.get(url, params = parameters)
    results = json.loads(res.content)
    return results['candidates'][0]['place_id']

inputtype = 'textquery'
```

D. Use the [*Place Details*](https://developers.google.com/places/web-service/details#PlaceDetailsResults) Method to find the Reviews of a Place in English and Spanish. In order to fulfill this item you need to:
* Create a *find_reviews* function with the appropiate parameters. It must return a JSON with the results. 
* Create a table with the following columns: Author Name, Language, Rating, Text, Time.

```python
def find_reviews(key, place_id, language):
    """In this function return a json  that is util for find reviews
    arguments:key, place_id, language-> the language that we want"""
    
    url = "https://maps.googleapis.com/maps/api/place/details/json"
    paramaters={
        'key': key,
        'place_id':place_id,
        'language':language
    }
    
    res = requests.get(url, params = paramaters)
    result = json.loads(res.content)
    return result

language = 'es, en'
```
